```
import pandas as pd
import numpy as np
```

```
url = 'Autolib.csv'
```

```
Autolib = pd.read_csv(url)
```

```
#Data understanding 
print (Autolib.info())
print (Autolib.describe())
```

```
 Autolib.isnull().sum()
```

```
Autolib.drop(['Displayed comment'],inplace=True , axis=1)
```

```
Autolib.drop(['Scheduled at'], axis=1,inplace=True)
```

```
Autolib.head()
```

```
Autolib.sort_values(by='year', ascending=1,inplace=True)
Autolib
```

```
#column for the cars
Autolib['Bluecar_status']=Autolib['Bluecar counter'].diff()
Autolib['Utilib_status'] = Autolib['Utilib counter'].diff()
Autolib['Utilib_1.4_status']= Autolib['Utilib 1.4 counter'].diff()
```

```
Autolib[(Autolib.Bluecar_status > 0)].groupby('hour')['hour'].count().sort_values(ascending=False).head(1)
```

```
#Combining columns for date relevancy.
Autolib['Time'] = Autolib['year'] + Autolib['month'] 
Autolib['day'] + Autolib['hour'] 
Autolib['minute']

Autolib['Date'] = pd.to_datetime(Autolib['Time'])
Autolib.head()
```

```
Autolib[['Bluecar_status','Bluecar counter',
'Utilib_status','Utilib counter',
'Utilib_1.4_status','Utilib 1.4 counter']]
```

```
#create a table for Paris is the only city
Paree = Autolib[Autolib.City=='Paris']
Paree.head()
```

```
#Determine the most popular hour for cars arrived
Paree[(Autolib.Bluecar_status)> 0].groupby('hour')['hour'].count().sort_values(ascending=False).head(1)
```

```
#Determine the most popular hour for cars departed
Paree[(Paree.Bluecar_status < 0 )].groupby('hour')['hour'].count().sort_values(ascending=True).head(1)
```
